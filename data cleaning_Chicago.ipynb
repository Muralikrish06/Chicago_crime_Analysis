```python
#This script will install the pandas module and read the excel file
import pandas as pd
import numpy as np 
CCA=pd.read_excel("C:/Users/murkr/Downloads/Crime_Data.xlsx")
display(CCA)
```

```python
#TO KNOW THE DATA TYPES
CCA.info()
```

```python
#TO KNOW THE SHAPE OF THE DATA AND TO KNOW THE NULL VALUES
print(CCA.shape)
CCA.isnull().sum()
```

```python
#TO DROP THE COLUMNS WHICH ARE NOT REQUIRED AND HAS MORE NULL VALUES
CCA.drop(columns="Ward",inplace=True)
CCA.drop(columns='Community Area',inplace=True)
```

```python
#TO DROP THE ROWS WHICH HAS NULL VALUES
CCA.dropna(subset="Latitude",axis=0,inplace=True)
CCA.dropna(subset="Longitude",axis=0,inplace=True)
CCA.dropna(subset="Location",axis=0,inplace=True)
CCA.dropna(subset='X Coordinate',axis=0,inplace=True)
CCA.dropna(subset='Y Coordinate',axis=0,inplace=True)
CCA.dropna(subset='Location Description',axis=0,inplace=True)
```

```python
#TO CONVERT THE DATE COLUMN INTO DATE TIME FORMAT AND TO CREATE NEW COLUMNS LIKE MONTH,TIME AND TIME_HOUR
CCA["Date"]=pd.to_datetime(CCA["Date"])
CCA["Month"]=CCA["Date"].dt.month
CCA["Time"]=CCA["Date"].dt.time
CCA["TIME_HOUR"]=CCA["Date"].dt.hour
CCA["Day"]=CCA["Date"].dt.day_name()
```

```python
#TO FIND THE DUPLICATES
duplicate_rows = CCA[CCA.duplicated()]
duplicate_rows
```

```python
#TO FIND THE DUPLICATES IN CASE NUMBER
Duplicate_CN=CCA["Case Number"].duplicated()
CCA[Duplicate_CN]
```

```python
#TO DROP THE DUPLICATES IN CASE NUMBER
CCA.drop_duplicates(subset=['Case Number'], inplace=True)
```

```python
#TO KNOW THE SHAPE AFTER DROPING THE DUPLICATES
CCA.shape
```

```python
#TO KNOW THE UNIQUE VALUES IN THE TIME_HOUR COLUMN
CCA["TIME_HOUR"].unique()
```

```python
#CREATE NEW COLUMN PART_OF_DAY DASED ON THE TIME_HOUR COLUMN
CCA["part_of_day"]=CCA["TIME_HOUR"].replace({6:"Morning",7:"Morning",8:"Morning",9:"Morning",10:"Morning",11:"Morning",12:"Noon",13:"Noon",14:"Noon",15:"Noon",16:"Evening",17:"Evening",18:"Night",19:"Night",20:"Night",21:"Night",22:"Night",23:"Night",24:"Night",1:"Night",2:"Night",3:"Night",4:"Night",5:"Night",0:"Night"})
CCA["part_of_day"].unique()
```

```python
#TO REPLACE THE MONTH COLUMN WITH THE MONTH NAMES
CCA["Month"]=CCA["Month"].replace({1:"JANUARY",2:"FEBRUARY",3:"MARCH",4:'APRIL',5:'MAY',6:'JUNE',7:'JULY',8:'AUGUST',9:'SEPTEMBER',10:'OCTOBER',11:'NOVEMBER',12:'DECEMBER'})
CCA["Month"].unique()
```

```python
#TO CREATE NEW COLUMN SEASON BASED ON THE MONTH COLUMN
CCA["Season"]=CCA["Month"].replace({"JANUARY":"Winter",'FEBRUARY':"Winter",'MARCH':"Spring",'APRIL':"Spring",'MAY':"Spring",'JUNE':"Summer",'JULY':"Summer",'AUGUST':"Summer",'SEPTEMBER':"Autumn",'OCTOBER':"Autumn",'NOVEMBER':"Autumn",'DECEMBER':"Winter"})
CCA["Season"].unique()
```

```python
CCA.rename(columns={"Primary Type":"Type_of_crime"},inplace=True)
CCA.rename(columns={"Location Description":"Description_of_location"},inplace=True)
CCA.rename(columns={"Case Number":"Case_Number"},inplace=True)
```

```python
import pandas as pd
import mysql.connector
from mysql.connector import Error
def create_table_and_insert(csv_file_path, table_name, host, user, password, database, batch_size):
    connection = None  # Initialize connection to None

    # Load the CSV into a Pandas DataFrame
    df = pd.read_csv(csv_file_path)

    # Establish a connection to the MySQL database
    connection = mysql.connector.connect(
        host=host,
        user=user,
        password=password,
        database=database
    )
    if connection.is_connected():
        print("Connected to MySQL database")
        cursor = connection.cursor()

        # Create the table based on the DataFrame columns
        create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ("
        for col in df.columns:
            # Use backticks to quote column names and handle reserved keywords
            # Also, define more appropriate data types 
                if df[col].dtype == 'object':
                # If it's a string, use VARCHAR(255)
                    create_table_query += f"`{col}` VARCHAR(255),"            
                elif df[col].dtype == 'int64':
                # If it's an integer, use INT
                    create_table_query += f"`{col}` INT,"
                elif df[col].dtype == 'float64':
                # If it's a float, use DOUBLE
                    create_table_query += f"`{col}` DOUBLE,"
                elif df[col].dtype == 'datetime64[ns]':
                # If it's a datetime, use DATETIME
                    create_table_query += f"`{col}` DATETIME,"
                else:
                 # For other types, use VARCHAR as a fallback
                    create_table_query += f"`{col}` VARCHAR(255),"
        create_table_query = create_table_query.rstrip(',') + ')'  # Remove the trailing comma

        cursor.execute(create_table_query)
        print(f"Table `{table_name}` created or already exists")

        # Prepare the data for insertion
        insert_query = f"INSERT INTO {table_name} ({', '.join([f'`{col}`' for col in df.columns])}) VALUES ({', '.join(['%s'] * len(df.columns))})"
        # Insert data in batches
        for start in range(0, len(df), batch_size):
            end = start + batch_size
            batch_data = df.iloc[start:end].values.tolist()
            cursor.executemany(insert_query, batch_data)
            connection.commit()
            print(f"Inserted rows {start+1} to {end} into `{table_name}`")

        print(f"Data inserted successfully in batches of {batch_size}")
# Usage
create_table_and_insert(
    csv_file_path="C:/Users/murkr/OneDrive/Desktop/CHICAGO_CRIME/clean_crime_data.csv",  # Path to the CSV file
    table_name="CCRIME",  # The name of the table you want to create
    host="gateway01.ap-southeast-1.prod.aws.tidbcloud.com",  # Your MySQL host
    user="32oPpe4fkuHmPwx.root",  # Your MySQL username
    password="O3ln75Jd20O0kbRa",  # Your MySQL password
    database="MURALI",  # Your MySQL database name
    batch_size=1000  # Batch size for inserting rows
)  
cursor.close()
connection.close()      



```
